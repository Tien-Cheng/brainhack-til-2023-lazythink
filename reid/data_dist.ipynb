# visualise the distribution of classes in the training set, validation set and the 2 sets together

```python
import glob
import matplotlib.pyplot as plt
import copy
import os
```

## get frequency of classes in training set and plot them

```python
train_dict = {}

for file in glob.glob("./train_labels/*.txt"):
    with open(file) as f:
        detections = f.readlines()
        for item in detections:  
            id = item.split(' ')[0]
            if id in train_dict:
                train_dict[id] += 1
            else:
                train_dict[id] = 1

print(train_dict)
```

```python
# Extract keys and values from the dictionary
sorted_keys = sorted(train_dict, key=train_dict.get, reverse=True)
sorted_values = [train_dict[key] for key in sorted_keys]

plt.figure(figsize=(40,20))
# Plotting the bar graph
plt.bar(sorted_keys, sorted_values)
plt.xlabel('Keys')
plt.ylabel('Frequency')
plt.title('Frequency of Keys')

# Display the plot
plt.show()
```

```python
# Sort the dictionary items based on values in descending order
sorted_items = sorted(train_dict.items(), key=lambda x: x[1], reverse=True)

# Print the top 20 keys with their counts
print("Top 20 keys with the highest counts:")
for key, value in sorted_items[:20]:
    print(key, ":", value)
```

```python
print(len(train_dict))
```

```python
print(sum(train_dict.values()))
```

## get frequency for validation set

```python
val_dict = {}
full_dict = copy.deepcopy(train_dict)

for file in glob.glob("./val_labels/*.txt"):
    with open(file) as f:
        detections = f.readlines()
        for item in detections:  
            id = item.split(' ')[0]
            if id in val_dict:
                val_dict[id] += 1
            else:
                val_dict[id] = 1
            if id in full_dict:
                full_dict[id] += 1
            else:
                full_dict[id] = 1

print(val_dict)
print(full_dict)
```

```python
# Extract keys and values from the dictionary
sorted_keys = sorted(val_dict, key=val_dict.get, reverse=True)
sorted_values = [val_dict[key] for key in sorted_keys]

plt.figure(figsize=(40,20))
# Plotting the bar graph
plt.bar(sorted_keys, sorted_values)
plt.xlabel('Keys')
plt.ylabel('Frequency')
plt.title('Frequency of Keys')

# Display the plot
plt.show()
```

```python
# Sort the dictionary items based on values in descending order
sorted_items = sorted(val_dict.items(), key=lambda x: x[1], reverse=True)

# Print the top 20 keys with their counts
print("Top 20 keys with the highest counts:")
for key, value in sorted_items[:20]:
    print(key, ":", value)
```
